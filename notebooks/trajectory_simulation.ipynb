## 2. Custom Trajectory Simulation

Simulate a patient progressing from drug-naive → responsive → DRE.


```python
engine = LPAEngine()
pid = 'SIM-DEMO-001'

records = [
    ('2020-01-01', 'TIER1', 'Focal impaired awareness seizures since age 14. MRI: left hippocampal sclerosis.'),
    ('2020-06-01', 'TIER1', 'Seizure-free on levetiracetam 1000mg BID. Good drug response.'),
    ('2022-03-01', 'TIER1', 'Breakthrough seizures 4x/month. Escalated to levetiracetam 3000mg. Added lacosamide.'),
    ('2023-08-01', 'TIER1', 'Failed levetiracetam, lacosamide, and oxcarbazepine. DRE confirmed. Surgical workup initiated.'),
]

for date_str, tier_str, text in records:
    rec = IngestionRecord(
        patient_id=pid,
        timestamp=datetime.fromisoformat(date_str),
        data_type=DataRecordType.NOTE,
        source_tier=SourceTier[tier_str],
        raw_text=text,
        author='simulation',
        version='1.0',
    )
    engine.ingest(rec)

print(f'Ingested {len(records)} records for {pid}')
```

```python
# Resolve CSEP
report = engine.resolve_csep(pid)
import json
print(json.dumps(engine.export_summary(pid), indent=2, default=str))
```
